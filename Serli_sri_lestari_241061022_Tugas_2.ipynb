# **1. Import Library & Load Data**

```python
# Import library yang digunakan
import pandas as pd
import numpy as np

# Load dataset (ganti dengan nama file kamu jika perlu)
df = pd.read_csv('/content/data_tugas 2.csv')

# Menampilkan 5 data pertama
df.head()
```

# **2. Mengecek Missing Value**

```python
# Mengecek jumlah data kosong di setiap kolom
missing_values = df.isnull().sum()

# Menampilkan hasil
print(missing_values)
```

**Penjelasan:**

- isnull() → mendeteksi nilai kosong
- sum() → menghitung jumlahnya per kolom

# **3. Mengatasi Missing Value (Kategori → "unknown")**

```python
# Daftar kolom kategori
kategori_cols = ['make', 'bahan-bakar', 'jumlah-pintu', 'body-style']

# Mengganti nilai kosong dengan "unknown"
df[kategori_cols] = df[kategori_cols].fillna('unknown')

# Menampilkan jumlah tiap kategori
for col in kategori_cols:
    print(f"\n{col}")
    print(df[col].value_counts())
```

**Penjelasan:**

- fillna('unknown') → mengganti data kosong
- value_counts() → menghitung jumlah kategori

# **4. Mengatasi Missing Value (Numerik → Mean)**

```python
# Daftar kolom numerik
numerik_cols = ['length', 'width', 'height']

# Mengisi missing value dengan rata-rata
for col in numerik_cols:
    mean_value = df[col].mean()  # menghitung rata-rata
    df[col].fillna(mean_value, inplace=True)  # mengganti nilai kosong
```

**Penjelasan:**

- mean() → menghitung nilai rata-rata
- digunakan karena cocok untuk data numerik

# **5. Min-Max Normalization**

```python
# Normalisasi Min-Max pada wheel-bases
df['wheel-bases-normalized'] = (
    (df['wheel-bases'] - df['wheel-bases'].min()) /
    (df['wheel-bases'].max() - df['wheel-bases'].min())
)
```

**Penjelasan:**

- Mengubah nilai ke rentang 0–1
- Berguna agar data seimbang saat analisis

# **6. Z-Score Normalization**

```python
# Normalisasi Z-Score
for col in numerik_cols:
    df[col + '_zscore'] = (
        (df[col] - df[col].mean()) / df[col].std()
    )
```

**Penjelasan:**

- Mengubah data berdasarkan mean dan standar deviasi
- Digunakan untuk deteksi outlier
