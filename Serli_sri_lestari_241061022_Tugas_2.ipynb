## **1. Import Library (Cell 1)**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

**Penjelasan:**

- numpy → untuk simulasi distribusi Poisson
- matplotlib → untuk membuat histogram
- pandas → untuk mengolah data hasil simulasi

## **2. Menentukan Parameter Poisson (Cell 2)**

Rata-rata pelanggan **5 orang setiap 10 menit**

Dalam **1 jam (60 menit):**

            5×6=30

Jadi** λ (lambda) = 30 pelanggan per jam**

```python
lambda_per_hour = 30
```

## **3. Simulasi Jumlah Pelanggan 100 Hari (Cell 3)**

Simulasi menggunakan distribusi Poisson.

```python
np.random.seed(42)

simulasi_pelanggan = np.random.poisson(lambda_per_hour, 100)

simulasi_pelanggan
```

**Penjelasan:**

- poisson() → menghasilkan data acak distribusi Poisson
- 100 → jumlah simulasi (100 hari)

## **4. Visualisasi Histogram (Cell 4)**

```python
plt.hist(simulasi_pelanggan, bins=10)
plt.title("Histogram Jumlah Pelanggan per Jam (Simulasi 100 Hari)")
plt.xlabel("Jumlah Pelanggan")
plt.ylabel("Frekuensi")
plt.show()
```

**Penjelasan:**

Histogram menunjukkan **distribusi jumlah pelanggan per jam** selama 100 hari simulasi.

## **5. Menghitung Rata-rata dan Variansi (Cell 5)**

```python
rata_rata = np.mean(simulasi_pelanggan)
variansi = np.var(simulasi_pelanggan)

print("Rata-rata pelanggan per jam:", rata_rata)
print("Variansi pelanggan per jam:", variansi)
```

**Penjelasan:**

- mean() → menghitung rata-rata pelanggan
- var() → menghitung variansi data

Kapasitas restoran = **40 pelanggan per jam**

Probabilitas **lebih dari 40 pelanggan**

```python
kelebihan = np.sum(simulasi_pelanggan > 40)

probabilitas = kelebihan / len(simulasi_pelanggan)

print("Jumlah hari melebihi kapasitas:", kelebihan)
print("Probabilitas kelebihan kapasitas:", probabilitas)
```
